```python
import pandas as pd
import numpy as np
import yfinance as yf
import talib
import matplotlib.pyplot as plt
```

```python
# Download data
assets = ['MSFT', 'LLY', 'CVX', 'AAPL', 'BHP']
data = {ticker: yf.download(tickers=ticker, period='6mo')['Adj Close'] for ticker in assets}
prices = pd.DataFrame(data)
```

```python
# Calculate returns
returns = prices.pct_change().dropna()
returns.head()
```

```python
# Calculate Moving Averages
moving_averages = pd.DataFrame(index=prices.index)
for ticker in assets:
    moving_averages[f'{ticker}_SMA_30'] = prices[ticker].rolling(window=30).mean()
    moving_averages[f'{ticker}_SMA_90'] = prices[ticker].rolling(window=90).mean()

# Calculate RSI
rsi = pd.DataFrame(index=prices.index)
for ticker in assets:
    rsi[ticker] = talib.RSI(prices[ticker])

# Calculate Bollinger Bands
bollinger_bands = pd.DataFrame(index=prices.index)
for ticker in assets:
    upper_band, middle_band, lower_band = talib.BBANDS(prices[ticker])
    bollinger_bands[f'{ticker}_UpperBand'] = upper_band
    bollinger_bands[f'{ticker}_MiddleBand'] = middle_band
    bollinger_bands[f'{ticker}_LowerBand'] = lower_band
```

```python
# Concatenate indicators with returns
features = pd.concat([returns, moving_averages, rsi, bollinger_bands], axis=1).dropna()
features.head()
```

```python
from sklearn.decomposition import PCA

# Apply PCA
pca = PCA(n_components=5)
principal_components = pca.fit_transform(features)

# Convert PCA components to a DataFrame
pca_df = pd.DataFrame(data=principal_components, columns=[f'PC{i+1}' for i in range(pca.n_components_)], index=features.index)

# Concatenate PCA components with original features
features_with_pca = pd.concat([features, pca_df], axis=1)
features_with_pca.head()
```

```python
# Adding lagged features
def add_lagged_features(df, lags=5):
    lagged_df = df.copy()
    for lag in range(1, lags + 1):
        lagged_df = pd.concat([lagged_df, df.shift(lag).add_suffix(f'_lag{lag}')], axis=1)
    return lagged_df

lagged_features = add_lagged_features(returns)
features_with_lags = pd.concat([features, lagged_features], axis=1).dropna()
features_with_lags.head()
```
